# Data cleaning of Customer call list dataset.
## Dataset Description
Customer Data collected by private companies on their website when individual sign up to get access of the website content. Let us assume a scenerio that this private company is our client and we provide data analytics services. Once colled this data needs to be cleaned and delivered to thier marketing team. 

```python
import pandas as pd
```

```python
df=pd.read_excel('CustomerCallList.xlsx')
```

```python
df.head()
```

## Feature Description
- CustomerID - (numerical): Unique customer ID.
- First_Name - (Mixed - String): First name of an individual customer.
- Last_Name - (Mixed - String): Last name of an individual customer.
- Phone_Number - (Mixed - String): Phone number of customer.
- Address - (Mixed - String): Address of customer.
- Paying Customer - (Categorical): Paid for service.
- Do_Not_Contact - (Categorical): Not want to get in contact.
- Not_Useful_Column - (Categorical): Usefulness

## Data Assessment
### Dirty Data Issue - Content problems in dataset.
- `Last_Name` column has special characters in the last name.
- `Phone_Number` has NA values.  
- `Address` has NA value.
- `Paying_Customer` has NA value.
- `Do_Not_Contact` has NA value.
### Messy Date Issue - Structural problems in dataset.
- `Useless` column has to be dropped.
- `Phone_Number` columns does not have common structure for numbers.
- `Address` does have common structure, splitting Street name, state and zipcode.
- `Paying_Customer` has Yes No Y and N.
- `Do_Not_Contact` similar problem as above.

```python
df
```

```python
# Checking for duplicates
df.duplicated()
```

```python
df[df.duplicated()==1]
```

```python
df=df.drop_duplicates()
```

```python
# Removing Useless columns
df=df.drop(columns='Not_Useful_Column')
df.head()
```

```python
# Cleaning Last name column using string strip
df['Last_Name']=df['Last_Name'].str.strip('123 ... / _') 
# Just mentioning the characters 
# that we want to strip otherwise it will only remove whitespaces.
```

```python
# Changing all the phone numbers into comman structure.
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df.head()
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
df
```

```python
# Removing Nan values
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
df
```

```python
# Splitting address into street , state and zipcode.
df[['Street_Address','State','Zip_Code']]=df['Address'].str.split(',',expand=True)
df.head()
```

```python
# Making Paying_Customer consitent
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
# Making Do_Not_Contact consitent
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
# replacing N/a with blank space
df=df.replace('N/a','')
df=df.fillna('')
df
```

```python
# Removing all the customers records which have Do_not_contact value as Y.
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)

df
```

```python
# Removing all the customers records which do not have phone number.
for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)

df
```

```python
# Since we have removed some customer records, individuals with missing number 
# and who has not given us consent to contact. 
# We have to reset index before delivering to client.
df=df.reset_index(drop=True)
df
```
